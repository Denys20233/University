5. Замініти комірки з пропущеними значенями на середні за стовпцем. І тип даних на float. Якщо є пропуски, замінии їх на числовий.

```python
numeric_columns = ['subscribers', 'video views', 'uploads', 'subscribers_for_last_30_days',
                   'country_rank', 'channel_type_rank', 'video_views_for_the_last_30_days',
                   'created_year', 'created_date', 'video_views_rank']

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

for col in numeric_columns:
    df[col] = df[col].fillna(df[col].mean())

text_columns = df.select_dtypes(include='object').columns
for col in text_columns:
    df[col] = df[col].fillna(df[col].mode()[0])

df[numeric_columns] = df[numeric_columns].astype(float)

df
```

```python
df = df.dropna() # залишилися 123 країни без необхідних даних
```

9. Визначте максимальну, мінімальну, середню кількість переглядів (використовуйте методи max, min, mean).

10. Знайдіть країну, де найбільше відео було завантажено на YouTube.
