8. Знайти мінімальну та максимальну кількість відвідувачів для кожного музею за 2018 рік

```python
df['month'] = pd.to_datetime(df['month'], format='%m/%d/%Y %I:%M:%S %p')

df_2018 = df[df['month'].dt.year == 2018]

min_visitors = df_2018.iloc[:, 1:].min()
max_visitors = df_2018.iloc[:, 1:].max()
result = pd.DataFrame({'min': min_visitors, 'max': max_visitors})

result
```

9. Визначити місяці з найвищою і найнижчою загальною кількістю відвідувачів серед усіх музеїв для 2015 року. Вивести назви місяців (не цифри).

```python
df_2015 = df[df['month'].dt.year == 2015].copy()

df_2015['total_visitors'] = df_2015.iloc[:, 1:].sum(axis=1)

min_row = df_2015.loc[df_2015['total_visitors'].idxmin()]
max_row = df_2015.loc[df_2015['total_visitors'].idxmax()]

min_month = min_row['month'].strftime('%B')
max_month = max_row['month'].strftime('%B')
print('Місяць з найнижчою загальною кількістю відвідувачів:', min_month)
print('Місяць з найвищою загальною кількістю відвідувачів:', max_month)
```

10. Порівняти кількість відвідувачів музею "Avila Adobe" у літні і зимові місяці 2018 року

```python
df_2018 = df[df['month'].dt.year == 2018].copy()

df_2018['month_num'] = df_2018['month'].dt.month

summer = df_2018[df_2018['month_num'].isin([6, 7, 8])]['avila_adobe'].sum()
winter = df_2018[df_2018['month_num'].isin([1, 2, 12])]['avila_adobe'].sum()

print(f"Влітку: {summer} людей")
print(f"Взимку: {winter} людей")
```

12. Лінійний графік (відвідуваність Avila Adobe з часом)


```python
sns.set_theme(style='ticks')
sns.lineplot(data=df, x='month', y='avila_adobe')
plt.title('Avila Adobe: Відвідуваність з часом')
plt.show()
```

13. Boxplot (розподіл відвідувачів Chinese American Museum по роках)

```python
df['month'] = pd.to_datetime(df['month'])
df['year'] = df['month'].dt.year

sns.set_theme(style='darkgrid')
plt.figure(figsize=(10, 6))
sns.boxplot(x='year', y='chinese_american_museum', hue='year', data=df, palette='muted', legend=False)
plt.title('Розподіл відвідувачів Chinese American Museum по роках')
plt.xlabel('Рік')
plt.ylabel('Кількість відвідувачів')
plt.grid(True)
plt.tight_layout()
plt.show()
```

14. Heatmap (кореляція між музеями)

```python
museum_cols = df.select_dtypes(include='number').drop(columns=['year'], errors='ignore')
corr = museum_cols.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap='magma', fmt=".2f", linewidths=0.5)
plt.title('Кореляція між музеями')
plt.tight_layout()
plt.show()
```

15. Histogram (розподіл кількості відвідувачів Gateway to Nature Center)

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='gateway_to_nature_center', bins=20, kde=False, color='maroon')
plt.title('Розподіл кількості відвідувачів Gateway to Nature Center')
plt.xlabel('Кількість відвідувачів')
plt.ylabel('Кількість місяців')
plt.tight_layout()
plt.show()
```

16. Barplot (середня кількість відвідувачів по музеях за всі роки)

```python
mean_visitors = df.drop(columns='month').mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=mean_visitors.values, y=mean_visitors.index, palette='PRGn')
plt.title('Середня кількість відвідувачів по музеях (усі роки)')
plt.xlabel('Середня кількість відвідувачів')
plt.ylabel('Музей')
plt.tight_layout()
plt.show()
```
