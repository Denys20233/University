# Перевести типи даних в числові

Стовпець 'Gender':

```python
df['Gender'].value_counts()
```

```python
df['Gender'] = df['Gender'].map({'Female': 0, 'Male': 1})
```

Стовпець 'Occupation':

```python
df['Occupation'].value_counts()
```

```python
occupation_map = {
    'Nurse': 0,
    'Doctor': 1,
    'Engineer': 2,
    'Lawyer': 3,
    'Teacher': 4,
    'Accountant': 5,
    'Salesperson': 6,
    'Scientist': 7,
    'Software Engineer': 8,
    'Sales Representative': 9,
    'Manager': 10
}

df['Occupation'] = df['Occupation'].map(occupation_map)
```

Стовпець 'BMI Category':


```python
df['BMI Category'].value_counts()
```

```python
df['BMI Category'] = df['BMI Category'].replace('Normal Weight', 'Normal')
```

```python
df['BMI Category'].value_counts()
```

```python
category_map = {
    'Normal': 0,
    'Overweight':	1,
    'Obese': 2
    }

df['BMI Category'] = df['BMI Category'].map(category_map)
```

Стовпець 'Blood Pressure' розділимо на 2 стовпці('Systolic Blood Pressure', 'Diastolic Blood Pressure') і зімінимо текстові значення:

```python
df[['Systolic Blood Pressure', 'Diastolic Blood Pressure']] = df['Blood Pressure'].str.split('/', expand=True).astype(int)
df.drop(columns=['Blood Pressure'], inplace=True)
```

```python
df['Sleep Disorder'].value_counts()
```

```python
df['Sleep Disorder'] = df['Sleep Disorder'].map({'Sleep Apnea': 0, 'Insomnia': 1})
```

```python
df.info()
```

Вивести оновлений датасет

```python
df
```

Ще раз перевірити наявність пропущених даних

```python
df.isnull().sum()
```

Замінити пропущені значення на моду (або середні)

```python
most_common = df['Sleep Disorder'].mode()[0]
df['Sleep Disorder'] = df['Sleep Disorder'].fillna(most_common)
```

Вивести статистичні характеристики датасета

```python
df.info()
```

Вивести матрицю кореляції

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='mako', fmt=".2f")
plt.title('Матриця кореляції')
plt.show()
```

Приклади 3 графіків: matplotlib


```python
plt.figure(figsize=(6, 4))
plt.scatter(df['Occupation'], df['Physical Activity Level'])
plt.title('Зв\'язок між професією і фізичною активністю')
plt.xlabel('Професія')
plt.ylabel('Рівень фізичної активності')
plt.show()
```

```python
colors = ['#0a1d2c', '#10394d', '#1e5d65', '#3a7d6e', '#66a176', '#a5c37a']

plt.figure(figsize=(6, 6))
plt.pie(df['Stress Level'].value_counts(), colors = colors,
        labels=df['Stress Level'].value_counts().index, autopct='%1.1f%%')
plt.title('Розподіл рівня стресу')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['Occupation'], edgecolor='black')
plt.title('')
plt.xlabel('Професія')
plt.ylabel('Кількість')
plt.show()
```

Приклади 3 графіків plotly

```python
fig1 = px.histogram(df, x='Sleep Duration', color='Gender',
                   title='Розподіл тривалості сну за статтю',
                   labels={'Sleep Duration': 'Тривалість сну', 'Gender': 'Стать'},
                   barmode='overlay', opacity=0.75)
fig1.show()
```

```python
fig2 = px.box(df, x='Occupation', y='Sleep Duration', title='Залежність тривалості сну від професії',
              color='Occupation', color_discrete_sequence=px.colors.sequential.haline)

fig2.show()
```

```python
fig3 = px.scatter(df, x='Sleep Duration', y='Daily Steps', color='Sleep Duration',
                  color_continuous_scale=px.colors.diverging.Portland,
                  title='Зв\'язок між кількостю кроків і тривалістю сну')
fig3.show()
```

Приклади 3 графіків seaborn


```python
sns.displot(df, x='Sleep Duration', hue='Stress Level', kind='kde',
            fill=True, palette='crest')

plt.title('Щільність тривалості сну за рівнем стресу')
plt.xlabel('Тривалість сну')
plt.ylabel('Щільність')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Occupation', order=df['Occupation'].value_counts().index, palette='crest')
plt.title('Кількість людей за професіями')
plt.xlabel('Професія')
plt.ylabel('Кількість')
plt.tight_layout()
plt.show()
```

```python
sns.relplot(data=df, x='Sleep Duration', y='Daily Steps',
            hue='Stress Level', palette='viridis')
plt.title('Тривалість сну vs Кількість кроків')
```

Приклад графіка altair

```python
chart = alt.Chart(df).mark_bar().encode(
    alt.X('Age', bin=True), y='count()',
    color=alt.Color('Gender', scale=alt.Scale(scheme='redyellowblue'))
).properties(title='Гістограма розподілу віку')

chart.show()
```
